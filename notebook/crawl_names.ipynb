```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Create an URL object
url = 'http://thptbinhphu.edu.vn/danh-sach-hoc-sinh-lop-10a1-nam-hoc-2021-2022.html'
# Create object page
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'lxml')
soup
```

```python
table = soup.find('table')
table
```

```python
headers = ['no.','id','name','dob','gender','note']
df = pd.DataFrame(columns = headers)
# Create a for loop to fill mydata
for idx, rrow in enumerate(table.find_all('tr')[1:]):
 rdata = rrow.find_all('td')
 row = [val.text for val in rdata]
 df.loc[idx] = row
```
